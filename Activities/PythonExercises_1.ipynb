**Text Analysis Tool:** Create a Python program that reads a text file and performs various analyses on its content. The program should: Count the number of words, sentences, and paragraphs. Calculate the average word length. Identify the most common words and their frequencies.


```python
from google.colab import files
# Upload the text file
uploaded = files.upload()

# Read the text file
filename = 'sample_text.txt'
with open(filename, 'r') as file:
    filetext = file.read()

#libraries
import re
from collections import Counter

#code
def num_of_words(textfile):
  words = textfile.split()
  return len(words)
print('Number of words:', num_of_words(filetext))

def sentences(textfile):
  sentences = re.split(r'[.!?]', textfile)
  return len(sentences) - 1
print('Number of sentences:', sentences(filetext))

def paragraphs(textfile):
  paragraphs = re.split(r'\n\s*\n', textfile)
  return len(paragraphs)
print('Number of paragraphs', paragraphs(filetext))

def avg_word_length(textfile):
    words = textfile.split()
    total_chars = 0
    total_words = len(words)  # Total words in the text

    for word in words:
        total_chars += len(word)

    return total_chars / total_words

print('Average word length:', avg_word_length(filetext))


def common_words_freq(textfile, num_of_words=3):
    words = textfile.split()
    word_counts = Counter(words)
    most_common_words = word_counts.most_common(num_of_words)

    return most_common_words

# Read the text file
filename = 'sample_text.txt'  # Change this to the path of your text file
with open(filename, 'r') as file:
    filetext = file.read()

most_common_words = common_words_freq(filetext)

print('Common words:')
for word, frequency in most_common_words:
    print(f'{word}\t{frequency}')
```

**Hangman Game:** Create a Python program that allows a user to play the game of Hangman. The program should randomly select a word from a predefined list, and the player must guess letters one at a time until they guess the word correctly or run out of attempts.

```python
import random

word_list = ['ASTRONOMY', 'ARTIFICIAL', 'HANGMAN', 'INTELLIGENCE', 'PROGRAMMING', 'STAR']

def select_word():
    return random.choice(word_list)

def hangman(word):
    word_completion = "_" * len(word)
    guessed_letters = []
    attempts = 6

    print("Play Hangman!")
    print(word_completion)
    print("\n")

    while attempts > 0:
        guess = input("Please guess a letter: ").upper()

        if guess in guessed_letters:
            print("You already guessed that letter.")
        elif guess not in word:
            attempts -= 1
            print("Wrong guess! Attempts left:", attempts)
            guessed_letters.append(guess)
            display_hangman(attempts)
        else:
            guessed_letters.append(guess)
            word_as_list = list(word_completion)
            indices = [i for i, letter in enumerate(word) if letter == guess]
            for index in indices:
                word_as_list[index] = guess
            word_completion = "".join(word_as_list)
            print(word_completion)

        if "_" not in word_completion:
            print("Congratulations! You guessed the word:", word)
            return

    print("You ran out of attempts. The word was:", word)

def display_hangman(attempts):
    hangman_display = [
        '''
           -----
           |   |
               |
               |
               |
               |
        ''',
        '''
           -----
           |   |
           O   |
               |
               |
               |
        ''',
        '''
           -----
           |   |
           O   |
           |   |
               |
               |
        ''',
        '''
           -----
           |   |
           O   |
          /|   |
               |
               |
        ''',
        '''
           -----
           |   |
           O   |
          /|\  |
               |
               |
        ''',
        '''
           -----
           |   |
           O   |
          /|\  |
          /    |
               |
        ''',
        '''
           -----
           |   |
           O   |
          /|\  |
          / \  |
               |
        '''
    ]
    print(hangman_display[6 - attempts])

if __name__ == "__main__":
    word_to_guess = select_word()
    hangman(word_to_guess)
```

**Number Guessing Game:** Create a program that generates a random number between 1 and 100, and then asks the user to guess the number. The program should provide hints to the user if the guessed number is greater or less than the target number, and keep asking for a new guess until the user guesses correctly.



```python
import random

def random_number():
    target = random.randint(1, 100)
    return target

def ask_number(message="Guess the number between 1 and 100: "):
    user_guess = int(input(message))
    return user_guess

def check_number(target, user_guess):
    if user_guess < target:
        print("Too low:( Try again.")
    elif user_guess > target:
        print("Too high:( Try again.")
    else:
        print("Congratulations:D You guessed!")

def main():
    target_number = random_number()
    attempts = 0

    while True:
        user_guess = ask_number()
        attempts += 1
        check_number(target_number, user_guess)
        if user_guess == target_number:
            print(f"It took you {attempts} attempts.")
            break

if __name__ == "__main__":
    main()
```
